### Инициализируем Dataset и DataLoader
> Используем переопределенный класс SegmentationDataset()
</br>Нормализуем изображения (с $\mu$ = 0 и $\sigma$ = 1)
</br>Далее создаем итераторы для обучающего и валидационного набора 

```python
from torch.utils.data import DataLoader
from torchvision import transforms, datasets

norm = transforms.Compose([
    transforms.ToTensor(),
    transforms.Normalize(
        mean=[.0, .0, .0], 
        std=[1., 1., 1.])

])

train_set = dataset_utils.Segmentation_Dataset('../Segmentation_Data/Train/', transforms=norm)
val_set = dataset_utils.Segmentation_Dataset('../Segmentation_Data/Val/', transforms=norm)

train_loader = DataLoader(
    dataset=train_set, 
    batch_size=batch_size, 
    shuffle=True, 
    num_workers=num_workers
)

val_loader = DataLoader(
    dataset=val_set, 
    batch_size=batch_size, 
    shuffle=False, 
    num_workers=num_workers
)
```

## 2. Сегментация
### Модель
> В качестве алгоритма для классификации использовалась упрощенная архитектура Unet с уменьшенным количеством фильтров в каждом слое и немного измененными параметрами свертки.</br>
Глобально: меньшее кол-во фильтров,так как сегментироваться будут не столь "абстрактные" или "сложные" обьекты, измененные параметры свертки - чтобы не кропать изображение во время конкатенации слоев.

```python
from torchsummary import summary
from unet import UNet

device = 'cuda' if torch.cuda.is_available() else 'cpu'
model = UNet().to(device)
summary(model, input_size=(3, IMG_H, IMG_W))
```

### Обучение и гиперпараметры.
> Функция потерь - binary crossentropy т.к. маски были бинарные.</br>
Оптимизатор - Adam</br>
Также использовался lr_scheduler для уменьшения learning rate

```python
import torch.optim as optim
from torch.optim import lr_scheduler

loss_fn = nn.BCEWithLogitsLoss()
optimizer = optim.Adam(model.parameters(), lr=1e-3)
scheduler = lr_scheduler.StepLR(optimizer, step_size=30, gamma=0.1)
num_epochs = 40
```

В качестве метрики качества сегментации был выбран Dice Coefficient:</br>
$$DC = \frac{2*|X \cap Y|}{|X|+|Y|}$$</br>
На самом деле функция потерь представляет из себя композицию BCE loss и Dice loss:</br>
$$loss = bce\times b_1 + dice \times (1-b_1)$$
где $$b_1 = 0.4$$ 
</br>
Почему так?

```python
from train import *

model, train_losses, val_losses = train_model(
    model=model,
    optimizer=optimizer,
    scheduler=scheduler,
    train_loader=train_loader,
    val_loader=val_loader,
    num_epochs=num_epochs
)
```

```python
plot_loss(
    train_loss=train_losses,
    valid_loss=val_losses
)
```

```python
import seaborn as sns
def plot_locations(locations, kde, levels, s=20, fill=False, alpha=1, xlim=[0, 160], ylim=[160]):
    """
    Displays "Location intensity" on the image grid of each item
    """
    palette = sns.color_palette('muted')
    for _, item in enumerate(locations):
        coordinates = np.asarray(locations[item])
        plot = sns.jointplot(coordinates[:,0], coordinates[:,1], joint_kws={'s': s}, kind='scatter', alpha=alpha, ylim=160, xlim=[0, 160], color=palette[0])
        if kde == True:
            plot.plot_joint(sns.kdeplot, color=palette[1], zorder=0, levels=levels, fill=fill, linewidth=10)
        plot.fig.suptitle(f'Distribution of {item} positions', fontsize='x-large')
        plot.fig.subplots_adjust(top=0.95)
```

```python
plot_locations(
    locations=locations,
    kde=True,
    levels=5,
    s=10,
    fill=True,
    alpha=0.9
)
```
